```python
primed_mod_list = list(set(map(lambda x:x.replace(" ","_"),primed_mod_list)))
primed_mod_list
```

```python
import requests
import pandas as pd
from datetime import datetime

def find_day_diff(test):
    tmp = []
    for i in range(len(test)-1):
        tmp.append(int(str(test[i+1]-test[i]).split(',')[0][:-5]))
        # tmp.append(str(test[i+1]-test[i]).split(',')[0])
    tmp.append(int(str(datetime.now().date()-test[i+1]).split(',')[0][:-5]))
    return tmp

def find_target_df_index(html_data):
    df = pd.read_html(html_data)
    for i in range(len(df)):
        if "PC OnlyDates" in df[i].columns:
            return df[i]
    print("Something went wrong")
    return False

def str2datetime(df,name="Prime_Mod"):
    text = df.iloc[0][0]
    text2 = df.iloc[0][2]
    dic = {name:[]}
    tmp = ''
    for i in range(len(text)):
        tmp+=text[i]
        if len(tmp)==10:
            dic[name].append(datetime.strptime(tmp, '%Y-%m-%d').date())
            tmp=''
    
    tmp = ''

    try:
        for i in range(len(text2)):
            tmp+=text2[i]
            if len(tmp)==10:
                dic[name].append(datetime.strptime(tmp, '%Y-%m-%d').date())
                tmp=''
    except TypeError as e:
        print(name,':',e)
    except:
        print('something wrong',name)
    tmp = sorted(dic[name])
    dic[name] = find_day_diff(tmp)
    return dic


```

```python
# num = 17
# name = primed_mod_list[num]
url = "https://warframe.fandom.com/wiki/"
# url += name
result = {}

for i in range(len(primed_mod_list)):   
    html_data = requests.get(url + primed_mod_list[i]).text
    df = find_target_df_index(html_data)

    df = str2datetime(df,primed_mod_list[i])
    
    result.update(df)
    
result = dict(sorted(result.items(),key=lambda x:x[1][len(x[1])-1],reverse=True))

result
```

```python
list_predict_name = list(map(lambda x:x.replace("_"," "),list(result.keys())))
list_predict_name
```

```python
list_predict_last_seen = [f"{i[len(i)-1]:,}" + " days ago" if i[len(i)-1] !=1 else "1 day ago" for i in result.values()]
list_predict_last_seen
```

```python
predicted_df = pd.DataFrame()
predicted_df["Rank"] = range(1,len(list_predict_name)+1)
predicted_df["Mod Name"] = list_predict_name
predicted_df["Last Seen"] = list_predict_last_seen
predicted_df = predicted_df.set_index(["Rank"])
predicted_df
```
